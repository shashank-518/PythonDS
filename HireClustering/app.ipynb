```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python

iris = load_iris()
```

```python
iris
```

```python
df = pd.DataFrame(data=iris['data'] , columns=iris['feature_names'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)
```

```python
data_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, )
```

```python
x_scaled = pca.fit_transform(data_scaled)
```

```python
x_scaled
```

```python
plt.scatter(x_scaled[:,0], x_scaled[:,1], c=iris['target'])
```

```python
import scipy.cluster.hierarchy as sc

plt.figure(figsize=(10,7))
plt.title('DendoGram')

sc.dendrogram(sc.linkage(x_scaled, method='ward'))
plt.title("Dendo")
plt.xlabel("Sample Index")
plt.ylabel("Ecludian Distane")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2,linkage='ward')
cluster.fit(x_scaled)
```

```python
plt.scatter(x_scaled[:,0],x_scaled[:,1], c=cluster.labels_)
```
