```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
x2 = [2,3,5,6,7]

plt.plot(x1,x2 , color="red")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
```

```python
x = [1,2,3,4,5]
x2 = [2,3,5,6,7]
x3 = [7,8,9,10,11]

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.plot(x,x2)
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x,x3)
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,x2)
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,x2)
plt.title("Plot 1")

```
